### Service layer

Cos’è il service layer? Dato che sistemi diversi hanno bisogno di dati diversi nel senso di formati etc, si è pensato che non fosse possibile continuamente adattare i dati. La soluzione è il **service layer,** ovvero un nuovo layer nel quale il sistema che sta per inviare deve per forza adattare i dati in un formato “standard” e poi quel dato passerà nel **service layer** dell’altro sistema che procederà a prendere il formato standard e convertirlo in quello originale dell’azienda. ## Databases

### *Whats a many to many relationship?*

A **many-to-many relationship** occurs when multiple records in one table are associated with multiple records in another table.v

### ***Junction table?***

A **junction table** is used to link two tables in a many-to-many relationship. It contains foreign keys that reference the primary keys of the two related tables.

### *Join clause?*

The **JOIN** clause in SQL is indeed used to combine rows from two or more tables based on a related column, typically a primary key in one table and a foreign key in another.

- When there is a direct relationship, the JOIN creates a new temporary result set that includes the linked values.
- If there is no direct relationship, a **CROSS JOIN** (or Cartesian product) can occur, resulting in every combination of rows from the two tables.

### ***What is normalization in databases, and why is it important?***

**Normalization** is the process of organizing data in a database to reduce redundancy and improve data integrity. It involves structuring the database in such a way that:

- Each piece of data is stored only once.
- Relationships between data are clearly defined.

**Importance of Normalization:**

- **Reduces Data Redundancy**: By eliminating duplicate data, it saves storage space and ensures consistency.
- **Improves Data Integrity**: Changes to data are made in one place, reducing the risk of inconsistencies.
- **Facilitates Efficient Queries**: A well-structured database can lead to faster query performance.

### *Different type of joins*

- **INNER JOIN**: Returns only the rows that have matching values in both tables. This is the most common type of join.
- **LEFT JOIN (or LEFT OUTER JOIN)**: Returns all rows from the left table and the matched rows from the right table. If there is no match, NULL values are returned for columns from the right table.
- **RIGHT JOIN (or RIGHT OUTER JOIN)**: Returns all rows from the right table and the matched rows from the left table. If there is no match, NULL values are returned for columns from the left table.
- **SEMI JOIN**: Returns rows from the first table where there is a match in the second table, but does not return any columns from the second table. It essentially filters the first table based on the existence of a match.

### Handle vs Socket

A Python socket is used for network communication, exchanging data with remote systems, while a file handle is used for local file access on the filesystem. 

**Sockets** require connection setup (e.g., bind, connect), are typically bi-directional, and may block unless set to non-blocking mode. 
**File** handles are simpler, used to open and interact with files directly, and are usually uni-directional (read or write). 

Sockets handle network-related errors, whereas file handles encounter filesystem-related issues. Additionally, sockets require explicit connection management, while files are opened and closed directly.

**The socket does not read all of the data when it is opened**
